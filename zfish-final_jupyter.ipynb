```python
mono_supp = fig5.loc[(fig5['mono_bino'] == 0.0)].iloc[:,11:19] #S5A
bino_supp = fig5.loc[(fig5['mono_bino'] == 1.0)].iloc[:,11:19] #S5A
S5A = fig5.iloc[:,11:19] #S5A
mono = fig5.loc[(fig5['mono_bino'] == 0.0)].iloc[:,2:10] # 5A
bino = fig5.loc[(fig5['mono_bino'] == 1.0)].iloc[:,2:10] #5A
```

```python
cnt = []
for i in range(len(num_neurons)-1):
    cnt.append(num_neurons.values[i][0])
cnt
arr = np.linspace(0,255,256)
```

## Plotting -- 5a, S5a, 5b

```python
# basic b&w heatmap
sns.set(rc = {'figure.figsize':(35,6)})
sns.heatmap(data, cmap="gray").set(title="Binocular Pt Types Correlations")
```

```python
# each group has own color

import warnings
warnings.filterwarnings("ignore")

# mono = fig5.loc[(fig5['mono_bino'] == 0.0)].iloc[:,2:10]
mono.rename(columns={ mono.columns[0]: "one1" }, inplace = True)
mono.rename(columns={ mono.columns[1]: "one2" }, inplace = True)
mono.rename(columns={ mono.columns[2]: "one3" }, inplace = True)
mono.rename(columns={ mono.columns[3]: "one4" }, inplace = True)
mono.rename(columns={ mono.columns[4]: "one5" }, inplace = True)
mono.rename(columns={ mono.columns[5]: "two3" }, inplace = True)
mono.rename(columns={ mono.columns[6]: "two4" }, inplace = True)
mono.rename(columns={ mono.columns[7]: "two5" }, inplace = True)
mono = mono.dropna()

kwargs = dict(linewidths=.1, cbar=False, square=False)
ax = sns.heatmap(mono.transpose(), color='gray', alpha=0.2, **kwargs)

# iterate ending:cmap pairs
cmaps = {'1': 'Blues', '2': 'Reds', '3': 'gray_r', '4': 'Greens', '5': 'Purples', '6': 'Wistia', '7': 'nipy_spectral',
        '8': 'CMRmap'}

for ending, cmap in cmaps.items():
    
    # create mask for given stimulus ending
    mask = mono.transpose().apply(
        lambda x: x if x.name.endswith(ending) else 0,
        result_type='broadcast',
        axis=1,
    ).eq(0)
    
    # plot masked heatmap on reusable ax
    
    sns.heatmap(mono.transpose(), mask=mask, cmap=cmap, ax=ax, **kwargs)
```
